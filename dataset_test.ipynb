```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from dataset import XinguDataset
from onehotencoding import OneHotEncoding
```

```python
encoder = OneHotEncoding(3)
ds_no_aug = XinguDataset('scenes_allbands', 'truth_masks', encoder, [4, 6, 7], range(1, 11), 256)
ds_aug = XinguDataset('scenes_allbands', 'truth_masks', encoder, [4, 6, 7], range(1, 11), 256, True)
```

```python
print(f'ds_no_aug: {len(ds_no_aug)} samples')
print(f'ds_aug: {len(ds_aug)} samples')
```

```python
image, mask = ds_aug.__getitem__(64)
mask = encoder.decode(mask)
print(f'mask shape: {mask.shape}')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax[0].imshow(image.transpose(1, 2, 0))
ax[1].imshow(mask)
```
